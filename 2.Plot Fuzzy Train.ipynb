```python
def plot_membership_functions():
    # Antecedents
    cross_border = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'cross_border')
    country_risk = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'country_risk')
    pep_involvement = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'pep_involvement')
    transaction_type = ctrl.Antecedent(np.arange(0, 3, 0.1), 'transaction_type')
    
    # Consequent
    risk = ctrl.Consequent(np.arange(0, 101, 1), 'risk')

    # Membership Functions for Antecedents
    pep_involvement['no'] = fuzz.gaussmf(pep_involvement.universe, mean=0.1, sigma=0.1)
    pep_involvement['yes'] = fuzz.gaussmf(pep_involvement.universe, mean=0.9, sigma=0.1)
    cross_border['low'] = fuzz.gaussmf(cross_border.universe, mean=0.2, sigma=0.1)
    cross_border['high'] = fuzz.gaussmf(cross_border.universe, mean=0.8, sigma=0.1)
    country_risk['low'] = fuzz.gaussmf(country_risk.universe, mean=0.2, sigma=0.1)
    country_risk['high'] = fuzz.gaussmf(country_risk.universe, mean=0.8, sigma=0.1)
    
    # Correcting Membership Function definitions for transaction_type
    transaction_type['crypto_transfer'] = fuzz.gaussmf(transaction_type.universe, mean=0, sigma=0.1)
    transaction_type['payment'] = fuzz.gaussmf(transaction_type.universe, mean=1, sigma=0.3)
    transaction_type['other'] = fuzz.gaussmf(transaction_type.universe, mean=2, sigma=0.1)

    # Membership Functions for Consequent
    risk['low'] = fuzz.trimf(risk.universe, [0, 0, 50])
    risk['medium'] = fuzz.trimf(risk.universe, [20, 50, 80])
    risk['high'] = fuzz.trimf(risk.universe, [60, 100, 100])

    # Plotting the membership functions
    pep_involvement.view()
    cross_border.view()
    country_risk.view()
    transaction_type.view()
    risk.view()

    # Show the plots
    plt.show()

plot_membership_functions()
```
